```python
import numpy as np
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
irisdata = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
sepallength=irisdata[0:,0:1]
sepallength_min, sepallength_max = sepallength.min(), sepallength.max()
sepallength = (sepallength-sepallength_min)/(sepallength_max-sepallength_min)
sepallength
```
